```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 4 * x[0]**2 - x[0] * x[1] + 7 - x[0] - x[1]

def gradient_f(x):
    df_dx1 = 8 * x[0] - x[1] - 1
    df_dx2 = -x[0] - 1
    return np.array([df_dx1, df_dx2])
```

```python
def gradient_descent(starting_point, learning_rate, steps):
    trajectory = [starting_point]

    for _ in range(steps):
        current_point = trajectory[-1]
        gradient = gradient_f(current_point)
        new_point = current_point - learning_rate * gradient
        trajectory.append(new_point)

    return np.array(trajectory)
```

```python
starting_point = np.array([0, 0]) 
learning_rate = 0.1
steps = 5
trajectory = gradient_descent(starting_point, learning_rate, steps)
```

```python
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f([X1, X2])

plt.figure(figsize=(8, 6))
plt.contour(X1, X2, Z, levels=30)
plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', color='red')
plt.title('Gradient Descent Trajectory for 5 Steps (Starting from 0, 0)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```

```python
import numpy as np

class BinaryCrossEntropy:
    @staticmethod
    def backward(y_true, y_pred):
        return -(y_true / y_pred - (1 - y_true) / (1 - y_pred))

class Sigmoid:
    @staticmethod
    def forward(x):
        return 1 / (1 + np.exp(-x))
    
    @staticmethod
    def backward(x):
        sig = Sigmoid.forward(x)
        return sig * (1 - sig)
    
class Tanh:
    @staticmethod
    def forward(x):
        return np.tanh(x)
    
    @staticmethod
    def backward(x):
        return 1 - np.tanh(x)**2
```

```python
class NeuralNetwork:
    def __init__(self):
        self.W1 = np.random.uniform(-2, 2, (3, 2))  
        self.b1 = np.random.uniform(-2, 2, (3, 1))  
        self.W2 = np.random.uniform(-2, 2, (1, 3))  
        self.b2 = np.random.uniform(-2, 2, (1, 1)) 

    def forward(self, X):
        self.Z1 = np.dot(self.W1, X) + self.b1
        self.A1 = Tanh.forward(self.Z1)
        self.Z2 = np.dot(self.W2, self.A1) + self.b2
        self.A2 = Sigmoid.forward(self.Z2)
        return self.A2

    def backward(self, X, y_true, learning_rate):
        m = X.shape[1]
               
        dZ2 = self.A2 - y_true
        dW2 = (1 / m) * np.dot(dZ2, self.A1.T)
        db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)
        
        dZ1 = np.dot(self.W2.T, dZ2) * Tanh.backward(self.Z1)
        dW1 = (1 / m) * np.dot(dZ1, X.T)
        db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)
        
        self.W2 -= learning_rate * dW2
        self.b2 -= learning_rate * db2
        self.W1 -= learning_rate * dW1
        self.b1 -= learning_rate * db1
```

```python
X_train = np.random.rand(100, 2)  
y_train = np.random.randint(0, 2, (100, 1))

batch_size = 10
indices = np.random.choice(X_train.shape[0], batch_size, replace=False)
X_batch = X_train[indices].T
y_batch = y_train[indices].T

nn = NeuralNetwork()
learning_rate = 0.1
```

```python
print("Initial Weights (W1):")
print(nn.W1)
print("\nInitial Biases (b1):")
print(nn.b1)
print("\nInitial Weights (W2):")
print(nn.W2)
print("\nInitial Biases (b2):")
print(nn.b2)

initial_predictions = nn.forward(X_batch)
initial_loss = -np.mean(y_batch * np.log(initial_predictions) + (1 - y_batch) * np.log(1 - initial_predictions))
print(f"\nInitial Loss: {initial_loss}")

nn.backward(X_batch, y_batch, learning_rate)

print("\nUpdated Weights (W1) after 1 iteration:")
print(nn.W1)
print("\nUpdated Biases (b1) after 1 iteration:")
print(nn.b1)
print("\nUpdated Weights (W2) after 1 iteration:")
print(nn.W2)
print("\nUpdated Biases (b2) after 1 iteration:")
print(nn.b2)

updated_predictions = nn.forward(X_batch)
updated_loss = -np.mean(y_batch * np.log(updated_predictions) + (1 - y_batch) * np.log(1 - updated_predictions))
print(f"\nLoss after 1 iteration: {updated_loss}")
```

```python
def relu(x):
    return np.maximum(0, x)

def linear(x):
    return x


def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def linear_derivative(x):
    return np.ones_like(x)
```

```python
def forward_pass(inputs, weights_hidden, weights_output, w_bypass):
    hidden_layer_input = np.dot(inputs, weights_hidden)
    hidden_layer_output = relu(hidden_layer_input)
    
    output_layer_input = np.dot(hidden_layer_output, weights_output)
    output_layer_output = linear(output_layer_input)
    
    bypass_input = hidden_layer_output[:, 0]  
    output_layer_output += w_bypass * bypass_input.reshape(-1, 1)
    
    return hidden_layer_output, output_layer_output


def backward_pass(inputs, hidden_output, output, labels, weights_hidden, weights_output, w_bypass, learning_rate):
    output_error = output - labels
    output_grad = output_error * linear_derivative(output)

    hidden_error = np.dot(output_grad, weights_output.T)
    bypass_grad = np.dot(output_grad.T, hidden_output[:, 0]).reshape(-1, 1)
    hidden_grad = hidden_error * relu_derivative(hidden_output)

    weights_output -= learning_rate * np.dot(hidden_output.T, output_grad)
    weights_hidden -= learning_rate * np.dot(inputs.T, hidden_grad)
    w_bypass -= learning_rate * bypass_grad.mean()

    return weights_hidden, weights_output, w_bypass


def calculate_mse(predictions, labels):
    return np.mean((predictions - labels) ** 2)
```

```python
inputs = np.array([[-1, 1], [2, -3], [0, 2], [-1, -4]])
labels = np.array([2.2, 5.7, -2.9, 1.8]).reshape(-1, 1)

np.random.seed(42)
weights_hidden = np.random.randn(2, 3)
weights_output = np.random.randn(3, 1)
w_bypass = np.random.randn(1)
```

```python
epochs = 100
learning_rate = 0.01
for epoch in range(epochs):
    
    hidden_output, output = forward_pass(inputs, weights_hidden, weights_output, w_bypass)
    
    
    mse = calculate_mse(output, labels)
    
    
    weights_hidden, weights_output, w_bypass = backward_pass(inputs, hidden_output, output, labels, weights_hidden, weights_output, w_bypass, learning_rate)
    
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: MSE = {mse}")

print("\nFinal Weights Hidden Layer:")
print(weights_hidden)
print("\nFinal Weights Output Layer:")
print(weights_output)
print("\nFinal Bypass Weight:")
print(w_bypass)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def linear(x):
    return x

def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python
input_size = 1
hidden_layer1_size = 4
hidden_layer2_size = 4
output_size = 1

weights_hidden1 = np.random.randn(input_size, hidden_layer1_size)
bias_hidden1 = np.zeros((1, hidden_layer1_size))

weights_hidden2 = np.random.randn(hidden_layer1_size, hidden_layer2_size)
bias_hidden2 = np.zeros((1, hidden_layer2_size))

weights_output = np.random.randn(hidden_layer2_size, output_size)
bias_output = np.zeros((1, output_size))
```

```python
learning_rate = 0.2
epochs = 10

for epoch in range(epochs):
    hidden1_output = sigmoid(np.dot(p_values, weights_hidden1) + bias_hidden1)
    
    hidden2_output = relu(np.dot(hidden1_output, weights_hidden2) + bias_hidden2)
    
    output = linear(np.dot(hidden2_output, weights_output) + bias_output)
    
    loss = mse_loss(target_values, output)
    
    print(f"Epoch {epoch+1}/{epochs} - Loss: {loss}")
    
    output_error = (output - target_values) / num_data_points
    output_delta = output_error
    
    hidden2_error = np.dot(output_delta, weights_output.T)
    hidden2_delta = hidden2_error * (hidden2_output > 0)
    
    hidden1_error = np.dot(hidden2_delta, weights_hidden2.T)
    hidden1_delta = hidden1_error * hidden1_output * (1 - hidden1_output)
    
    weights_output -= learning_rate * np.dot(hidden2_output.T, output_delta)
    bias_output -= learning_rate * np.sum(output_delta, axis=0, keepdims=True)
    
    weights_hidden2 -= learning_rate * np.dot(hidden1_output.T, hidden2_delta)
    bias_hidden2 -= learning_rate * np.sum(hidden2_delta, axis=0, keepdims=True)
    
    weights_hidden1 -= learning_rate * np.dot(p_values.T, hidden1_delta)
    bias_hidden1 -= learning_rate * np.sum(hidden1_delta, axis=0, keepdims=True)
```

```python
learning_rate = 0.01
epochs = 10

for epoch in range(epochs):
    hidden1_output = sigmoid(np.dot(p_values, weights_hidden1) + bias_hidden1)
    
    hidden2_output = relu(np.dot(hidden1_output, weights_hidden2) + bias_hidden2)
    
    output = linear(np.dot(hidden2_output, weights_output) + bias_output)
    
    loss = mse_loss(target_values, output)
    
    print(f"Epoch {epoch+1}/{epochs} - Loss: {loss}")
    
    output_error = (output - target_values) / num_data_points
    output_delta = output_error
    
    hidden2_error = np.dot(output_delta, weights_output.T)
    hidden2_delta = hidden2_error * (hidden2_output > 0)
    
    hidden1_error = np.dot(hidden2_delta, weights_hidden2.T)
    hidden1_delta = hidden1_error * hidden1_output * (1 - hidden1_output)
    
    weights_output -= learning_rate * np.dot(hidden2_output.T, output_delta)
    bias_output -= learning_rate * np.sum(output_delta, axis=0, keepdims=True)
    
    weights_hidden2 -= learning_rate * np.dot(hidden1_output.T, hidden2_delta)
    bias_hidden2 -= learning_rate * np.sum(hidden2_delta, axis=0, keepdims=True)
    
    weights_hidden1 -= learning_rate * np.dot(p_values.T, hidden1_delta)
    bias_hidden1 -= learning_rate * np.sum(hidden1_delta, axis=0, keepdims=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

train_data = pd.read_csv('D:/datasets/MNIST/mnist_train.csv')
test_data = pd.read_csv('D:/datasets/MNIST/mnist_test.csv')

X_train = train_data.iloc[:, 1:].values / 255.0  
y_train = train_data.iloc[:, 0].values
X_test = test_data.iloc[:, 1:].values / 255.0
y_test = test_data.iloc[:, 0].values
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

input_size = X_train.shape[1]
output_size = 10  
weights = np.random.randn(input_size, output_size) * 0.01
biases = np.zeros(output_size)

def forward_pass(X):
    logits = np.dot(X, weights) + biases
    return softmax(logits)
```

```python
learning_rate = 0.1
num_epochs = 100
train_accuracy = []
for epoch in range(num_epochs):
    output = forward_pass(X_train)
    
    num_examples = len(X_train)
    corect_logprobs = -np.log(output[range(num_examples), y_train])
    loss = np.sum(corect_logprobs) / num_examples
    
    predictions = np.argmax(output, axis=1)
    accuracy = np.mean(predictions == y_train)
    train_accuracy.append(accuracy)
    
    d_logits = output
    d_logits[range(num_examples), y_train] -= 1
    d_logits /= num_examples
    
    d_weights = np.dot(X_train.T, d_logits)
    d_biases = np.sum(d_logits, axis=0)
    
    # Update weights and biases
    weights -= learning_rate * d_weights
    biases -= learning_rate * d_biases
```

```python
plt.plot(range(num_epochs), train_accuracy)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training Accuracy')
plt.show()
```

```python
test_output = forward_pass(X_test)
test_predictions = np.argmax(test_output, axis=1)
test_accuracy = np.mean(test_predictions == y_test)
print(f"Accuracy on test data: {test_accuracy * 100:.2f}%")
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.square(x)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x <= 0, 0, 1)

def linear(x):
    return x

def linear_derivative(x):
    return np.ones_like(x)

def binary_cross_entropy(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, activation_function, derivative_function):
        self.weights_input_hidden = np.random.randn(input_size, hidden_size) * np.sqrt(2. / input_size)
        self.weights_hidden_output = np.random.randn(hidden_size, output_size) * np.sqrt(2. / hidden_size)
        self.bias_hidden = np.zeros((1, hidden_size))
        self.bias_output = np.zeros((1, output_size))
        self.activation_function = activation_function
        self.derivative_function = derivative_function

    def forward_propagation(self, inputs):
        self.hidden_layer_input = np.dot(inputs, self.weights_input_hidden) + self.bias_hidden
        self.hidden_layer_output = self.activation_function(self.hidden_layer_input)
        self.output_layer_input = np.dot(self.hidden_layer_output, self.weights_hidden_output) + self.bias_output
        self.predicted_output = sigmoid(self.output_layer_input)
        return self.predicted_output

    def backward_propagation(self, inputs, outputs, learning_rate):
        error = outputs - self.predicted_output

        output_delta = error * sigmoid_derivative(self.predicted_output)
        hidden_error = output_delta.dot(self.weights_hidden_output.T)
        hidden_delta = hidden_error * self.derivative_function(self.hidden_layer_output)

        self.weights_hidden_output += self.hidden_layer_output.T.dot(output_delta) * learning_rate
        self.weights_input_hidden += inputs.T.dot(hidden_delta) * learning_rate
        self.bias_output += np.sum(output_delta) * learning_rate
        self.bias_hidden += np.sum(hidden_delta) * learning_rate

    def train(self, inputs, outputs, epochs, learning_rate):
        loss_history = []

        for epoch in range(epochs):
            predicted_output = self.forward_propagation(inputs)
            loss = binary_cross_entropy(outputs, predicted_output)
            loss_history.append(loss)

            self.backward_propagation(inputs, outputs, learning_rate)

            if epoch % 100 == 0:
                print(f"Epoch {epoch}: Loss {loss}")

        return loss_history
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def binary_cross_entropy(y_true, y_pred):
    return -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
np.random.seed(42)
input_neurons = 2
hidden_neurons = 2
output_neurons = 1

weights_hidden = np.random.randn(input_neurons, hidden_neurons)
bias_hidden = np.zeros((1, hidden_neurons))
weights_output = np.random.randn(hidden_neurons, output_neurons)
bias_output = np.zeros((1, output_neurons))
```

```python
learning_rate = 0.01
epochs = 10

for epoch in range(epochs):
    total_loss = 0

    for i in range(len(data_np)):
        input_data = data_np[i].reshape(1, -1)
        hidden_layer_input = np.dot(input_data, weights_hidden) + bias_hidden
        hidden_layer_output = relu(hidden_layer_input)
        output_layer_input = np.dot(hidden_layer_output, weights_output) + bias_output
        output = sigmoid(output_layer_input)

        loss = binary_cross_entropy(labels_np[i], output)
        total_loss += loss

        output_error = output - labels_np[i]
        output_delta = output_error * sigmoid_derivative(output_layer_input)
        hidden_error = np.dot(output_delta, weights_output.T)
        hidden_delta = hidden_error * relu_derivative(hidden_layer_input)

        weights_output -= learning_rate * np.dot(hidden_layer_output.T, output_delta)
        bias_output -= learning_rate * np.sum(output_delta, axis=0, keepdims=True)
        weights_hidden -= learning_rate * np.dot(input_data.T, hidden_delta)
        bias_hidden -= learning_rate * np.sum(hidden_delta, axis=0, keepdims=True)

    avg_loss = total_loss / len(data_np)
    print(f"Epoch {epoch + 1}/{epochs} - Loss: {avg_loss[0]}")
```
